## ASSIGNING OBJECTIDS

```python
# Function to assign unique ObjectIds based on year and country
def assign_object_ids(df):
    df['ObjectId'] = df['year'].astype(str) + '_' + df['country'].astype(str)
    df['ObjectId'] = df['ObjectId'].apply(hash) % (10**9)  # Ensure positive, unique integers
    return df

# Assign unique ObjectIds based on year and country and make 'ObjectId' the first column
water_use = assign_object_ids(water_use)
water_use = water_use[['ObjectId'] + [col for col in water_use if col != 'ObjectId']]

forest_df = assign_object_ids(forest_df)
forest_df = forest_df[['ObjectId'] + [col for col in forest_df if col != 'ObjectId']]

land_data = assign_object_ids(land_data)
land_data = land_data[['ObjectId'] + [col for col in land_data if col != 'ObjectId']]

# CLIMATE CHANGE

as_temp_chan_melted = assign_object_ids(as_temp_chan_melted)
as_temp_chan_melted = as_temp_chan_melted[['ObjectId'] + [col for col in as_temp_chan_melted if col != 'ObjectId']]

disaster_freq_melted = assign_object_ids(disaster_freq_melted)
disaster_freq_melted = disaster_freq_melted[['ObjectId'] + [col for col in disaster_freq_melted if col != 'ObjectId']]

forest_C_melted = assign_object_ids(forest_C_melted)
forest_C_melted = forest_C_melted[['ObjectId'] + [col for col in forest_C_melted if col != 'ObjectId']]

lan_cov_melted = assign_object_ids(lan_cov_melted)
lan_cov_melted = lan_cov_melted[['ObjectId'] + [col for col in lan_cov_melted if col != 'ObjectId']]
```

```python
def sanitize_variable_name(variable_name):
    # Replace spaces and special characters with underscores
    sanitized_name = ''.join(c if c.isalnum() or c == '_' else '_' for c in variable_name)
    return sanitized_name

# Function to separate and save by variable column
def separate_and_save_by_variable(df, variable_column, output_dir='separated_data'):
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Get a list of unique values in the specified variable column
    unique_variables = df[variable_column].unique()
    
    # Loop through unique variables
    for variable in unique_variables:
        # Sanitize the variable name
        sanitized_variable = sanitize_variable_name(variable)
        
        # Create a folder for each variable within the output_dir
        variable_output_dir = os.path.join(output_dir, sanitized_variable)
        os.makedirs(variable_output_dir, exist_ok=True)
        
        # Filter the DataFrame by the current variable
        filtered_df = df[df[variable_column] == variable]
        
        # Define the output file paths for CSV and JSON within the variable folder
        csv_output_file = os.path.join(variable_output_dir, f'{sanitized_variable}.csv')
        json_output_file = os.path.join(variable_output_dir, f'{sanitized_variable}.json')
        
        # Make 'ObjectId' the first column in the DataFrame
        filtered_df = filtered_df[['ObjectId'] + [col for col in filtered_df if col != 'ObjectId']]
        
        # Save the filtered DataFrame to a new CSV file within the variable folder
        filtered_df.to_csv(csv_output_file, index=False)
        
        # Save the filtered DataFrame to a new JSON file within the variable folder
        filtered_df.to_json(json_output_file, orient='records', lines=True)

# Example usage for water_use DataFrame
separate_and_save_by_variable(water_use, 'h20_variable', 'water_use_separated')

# Example usage for land_data DataFrame
separate_and_save_by_variable(land_data, 'land_variable', 'land_data_separated')

# Example usage for forest_df DataFrame
separate_and_save_by_variable(forest_df, 'forest_variable', 'forest_data_separated')
```
