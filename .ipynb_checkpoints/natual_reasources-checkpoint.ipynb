```python
# Read the CSV file into a DataFrame
water_use_df = pd.read_csv('C:/Users/kayvb/OneDrive/Desktop/Project-3-ChangeYoClimate-MKSD-/data/WATER_USE_21092023033150254.csv')
```

```python
# Function to determine the continent based on the country
def determine_continent(country):
    # Check if the 'country' belongs to the list of European countries
    if country in ['Switzerland', 'Czech Republic', 'Luxembourg', 'Russia', 'Norway', 'Greece', 'Italy', 'Austria', 'Belgium', 'Netherlands', 'Spain', 'Iceland', 'Germany', 'United Kingdom'
                     'Latvia', 'Romania', 'Bulgaria', 'Holy See (Vatican City)', 'San Marino', 'France', 'Finland', 'Slovenia', 'Sweden', 'Liechtenstein', 'Bosnia and Herzegovina', 'Monaco', 'Andorra',
                     'Faeroe Islands', 'Gibraltar', 'Macau, China']:
        # Return the continent as 'Europe' if the country is in the European list
        return 'Europe'
    # Check if the 'country' belongs to the list of North American countries
    elif country in ['Mexico', 'Canada', 'United States', 'Guadeloupe', 'Saint Lucia', 'Turks and Caicos Islands', 'Puerto Rico', 'Dominican Republic', 'Haiti', 'Saint Vincent and the Grenadines',
                     'British Virgin Islands', 'Saint Kitts and Nevis', 'Guatemala', 'Anguilla', 'Martinique', 'Aruba', 'Barbados', 'Greenland']:
         # Return the continent as 'North America' if the country is in the North American list
        return 'North America'
        # Check if the 'country' belongs to the list of Oceanian countries
    elif country in ['Australia', 'Kiribati', 'Solomon Islands', 'American Samoa', 'Fiji', 'Papua New Guinea', 'Tuvalu', 'Vanuatu', 'Nauru', 'Tokelau']:
        # Return the continent as 'Oceania' if the country is in the Oceanian list
        return 'Oceania'
        # Check if the 'country' belongs to the list of Asian countries
    elif country in ['India', 'Türkiye', 'Russia', 'Greece', "China (People's Republic of)", 'Indonesia', 'Korea', 'Estonia', 'Poland', 'Japan', 'Belgium',
                     'Netherlands', 'Iceland', 'Germany', 'United Kingdom', 'Kazakhstan', 'Bulgaria', 'Armenia', 'Georgia', 'Tajikistan', 'Azerbaijan', 'Kiribati',
                     'Holy See', 'Sri Lanka', 'Singapore', 'New Caledonia', 'Tuvalu', 'Bahrain', 'Cyprus', 'Armenia', 'Georgia', 'Kazakhstan', 'Iran', 'Syrian Arab Republic',
                     'Viet Nam', 'Kyrgyzstan', 'Afghanistan', 'Myanmar', 'Hong Kong, China', 'Taiwan', 'Uzbekistan', 'Maldives', 'Oman', 'Yemen', 'Cambodia', 'Bhutan', 'Philippines',
                     'Mongolia', 'Macau, China']:
         # Return the continent as 'Asia' if the country is in the Asian list
        return 'Asia'
        # Check if the 'country' belongs to the list of South American countries
    elif country in ['Costa Rica', 'Colombia', 'Brazil', 'Chile', 'Argentina', 'Peru', 'Guyana', 'Suriname', 'Bolivia', 'Paraguay', 'Uruguay', 'Falkland Islands (Malvinas)']:
        # Return the continent as 'South America' if the country is in the South American list
        return 'South America'
         # Check if the 'country' belongs to the list of African countries
    elif country in ['South Africa', 'Nigeria', 'Tunisia', 'Algeria', 'Ghana', 'Zimbabwe', 'Uganda', 'Burkina Faso', 'Guinea', 'Malawi', 'Namibia', 'Botswana', 'Cabo Verde', 'Ethiopia',
                     'Djibouti', 'Eswatini', 'Togo', 'Jamaica', 'Papua New Guinea', 'El Salvador', 'Botswana', 'Burundi', 'Angola', 'Madagascar', 'Congo', 'Tanzania', 'Kenya', 'Mauritius',
                     'Gambia', 'Morocco', 'Niger', 'Egypt', 'Comoros', 'Sudan', 'Mali', 'Sierra Leone', 'Benin', 'Liberia', 'Seychelles', 'Mauritania', 'Zambia', 'Chad', 'Mozambique']:
        # Return the continent as 'Africa' if the country is in the African list
        return 'Africa'
```

```python
# Define the calculate_and_reorder_statistics function (if not defined already)
def calculate_and_reorder_statistics(input_df, column_order):
    # Group the filtered data and calculate statistics
    statistics_df = input_df.groupby(['country', input_df.columns[1]])[input_df.columns[2]].agg(['mean', 'median', 'std', 'min', 'max', 'count']).reset_index()
    
    # Reorder columns based on the specified column order
    statistics_df = statistics_df[column_order]
    
    return statistics_df
```

```python
# Define a list of columns to drop by name
columns_to_drop = ['COU', 'VAR', 'YEA', 'Unit Code', 'Reference Period Code', 'Reference Period', 'Flag Codes']

# Drop the specified columns
water_use = water_use_df.drop(columns=columns_to_drop)

# Handle missing values (replace NaN with NULL)
water_use.where(pd.notna(water_use), None)
```

```python
# Define custom column names in the order of your DataFrame
column_rename = {
    'Country': 'country',
    'Variable': 'h20_variable',
    'Year': 'year',
    'Unit': 'h20_unit',
    'PowerCode Code': 'h20_powercode_code',
    'PowerCode': 'h20_powercode',
    'Value': 'h20_value',
    'Flags': 'h20_flags'
}

# Rename columns using the rename method and inplace=True
water_use.rename(columns=column_rename, inplace=True)

# Adding the 'Continent' column to the DataFrame
water_use.insert(1, 'continent', water_use['country'].apply(determine_continent))

# Check the DataFrame to see the renamed columns
(water_use)
```

```python
# Save the cleaned data to a new CSV file
water_use.to_csv('cleaned_h20_use.csv', index=False)
```

```python
# Clean the 'h20_value' column by converting it to numeric, and set non-convertible values to NaN
water_use['h20_value'] = pd.to_numeric(water_use['h20_value'], errors='coerce')

# Group by 'country' and calculate statistics for 'h20_value'
stats_h20 = water_use.groupby('country')['h20_value'].agg(['mean', 'median', 'std', 'min', 'max', 'count']).reset_index()

# Include the 'h20_variable' column in the resulting DataFrame
stats_h20['h20_variable'] = water_use.groupby('country')['h20_variable'].first().values

# Reorder columns as 'country', 'forest_variable', and then the statistical analysis columns
column_order = ['country', 'h20_variable', 'mean', 'median', 'std', 'min', 'max', 'count']

# Reorder the DataFrame columns based on the specified order
stats_h20 = stats_h20[column_order]

# Save the stat data to a new CSV file
stats_h20.to_csv('stats_h20.csv', index=False)

# Print the calculated statistics by continent
print('water_use Statistical Values by Country:')
(stats_h20)
```

```python
# Group the DataFrame by 'continent' and calculate statistics
w_statistics_by_continent = water_use.groupby('continent')['h20_value'].agg(['mean', 'median', 'std', 'min', 'max', 'count'])

# Reset the index to have a clean DataFrame
w_statistics_by_continent.reset_index(inplace=True)

# Save the cleaned data to a new CSV file
w_statistics_by_continent.to_csv('h20_continent_stats.csv', index=False)

# Print the calculated statistics by continent
print('water_use Statistical Values by Continent:')
(w_statistics_by_continent)
```
