```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
import numpy as np

#file path
forest_land_path = "C:\\Users\\zombi\\Project-3-ChangeYoClimate-MKSD-\\data\\clean_data\\natural_reasources\\land_data\\Forest___land_area.csv"

#read csv file into dataframe
forest_land_csv = pd.read_csv(forest_land_path)
```

```python
forest_land_csv
```

```python
columns = forest_land_csv[['continent', 'year', 'land_value']]

columns
```

```python
average_continent_year = forest_land_csv.groupby(['continent', 'year'])['land_value'].mean().reset_index()

average_continent_year
```

```python
continent = 'Africa'

continent_data = average_continent_year[average_continent_year['continent'] == continent]

plt.figure(figsize=(10, 6))
plt.bar(continent_data['year'], continent_data['land_value'], color='teal', alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Average Land Percent')
plt.title('Average Land Percent over Years in Africa')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Average_Land_Percent_over_Years_Africa.png")
plt.show()
```

```python
world = forest_land_csv[forest_land_csv['country'] == 'World']
average_world = world.groupby('year')['land_value'].mean().reset_index()
average_world
```

```python
# world
plt.figure(figsize=(12, 6))
plt.bar(average_world['year'], average_world['land_value'], color='teal', alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Average Land Percent')
plt.title('Average Land Value over Years in World')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Average_Land_Percent_over_Years_World.png")
plt.show()
```

```python
tenyears = ['1992-2001', '2002-2011', '2012-2021']

interval_averages = []

for interval in tenyears:
    start_year, end_year = map(int, interval.split('-'))
    interval_values = average_world[
        (average_world['year'] >= start_year) &
        (average_world['year'] <= end_year)
    ]
    average_value = interval_values['land_value'].mean()
    interval_averages.append(average_value)
    
plt.figure(figsize=(10, 6))

plt.bar(tenyears, interval_averages, color=['lightcoral', 'seagreen', 'steelblue'], alpha=0.6)
    
plt.xlabel('Year')
plt.ylabel('Average Land Percent')
plt.title('Average Land Percent for World in 10-Year Intervals')
#plt.legend([f'Interval {i+1}' for i in range(len(tenyears))], title='Intervals', loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Average_Land_Percent_over_Years_World_10-Year.png")
plt.show()
```

```python
continent = 'Africa'
continent_data = average_continent_year[average_continent_year['continent'] == continent]

plt.figure(figsize=(10,6))
plt.scatter(continent_data['year'], continent_data['land_value'], color='forestgreen', alpha=0.6, label='Data Points')

x_values = continent_data['year'].values
y_values = continent_data['land_value'].values
slope, intercept, r_value, p_value, std_err = stats.linregress(x_values, y_values)
line = slope * x_values + intercept

plt.plot(x_values, line, color='red', label=f'Linear Regression (r-squared{r_value**2:.2f})')
plt.xlabel('Year')
plt.ylabel('Average Land Percent for Africa')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("Average_Land_Percent_over_Years_World_AfricaScat.png")
plt.show()
```
