```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
import numpy as np

#file path
sea_path = "Data\\climate_change\\chang_n_mean_sea.csv"

#read csv file into dataframe
sea_csv = pd.read_csv(sea_path)
```

```python
sea_csv
```

```python
sea_group = sea_csv.groupby(['Measure', 'Year'])['Value'].mean().reset_index()
sea_group
```

```python
sea = 'Adriatic Sea'

subset = sea_group[sea_group['Measure'] == sea]

colors = plt.cm.get_cmap('tab10', 1)

plt.figure(figsize=(10,6))

plt.scatter(subset['Year'], subset['Value'], label=sea, color=colors(0))

slope, intercept, r_value, p_value, std_err = linregress(subset['Year'], subset['Value'])

plt.plot(subset['Year'], intercept + slope * subset['Year'], color=colors(0), linestyle='--', label=f'Lin. Reg. {sea}')

plt.xlabel('Year')
plt.ylabel('Average Value')
plt.title(f'Scatter Plot with Linear Regression for {sea}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = 'Caribbean Sea'
sea_data = sea_csv[sea_csv['Measure'] == data]
average_caribbean = sea_data.groupby('Year')['Value'].mean()
```

```python
sea = 'Caribbean Ocean'

years = average_caribbean.index
values = average_caribbean.values

colors = plt.cm.get_cmap('tab10', 1)

plt.figure(figsize=(10,6))

plt.scatter(years, values, label="Carribbean Sea Data", color=colors(0))

slope, intercept, r_value, p_value, std_err = linregress(years, values)

regression_line = slope * years + intercept

plt.plot(years, regression_line, color=colors(0), linestyle='--', label=f'Linear Regression Line (R-squared = {r_value**2:.2f})')

plt.xlabel('Year')
plt.ylabel('Average Value')
plt.title(f'Scatter Plot with Linear Regression for {sea}')
plt.legend()
plt.grid(True)
plt.show()
```
