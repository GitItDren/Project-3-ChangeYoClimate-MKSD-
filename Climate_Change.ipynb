```python
def reshape_and_save(df, id_columns, columns_to_drop, melted_df_name):
                     # output_csv_filename):
    """
    Reshape a DataFrame by melting it to have years in a single column and save it to a CSV file.

    Args:
    df (pd.DataFrame): The input DataFrame.
    id_columns (list): A list of column names to keep as identifiers.
    columns_to_drop (list): A list of column names to drop.
    melted_df_name (str): Name for the melted DataFrame.
    output_csv_filename (str): Filename for the output CSV file.

    Returns:
    pd.DataFrame: The reshaped DataFrame.
    """
    # Drop specified columns
    df = df.drop(columns=columns_to_drop)

    # Replace missing values with NULL
    df = df.where(pd.notna(df), None)

    # Reshape the DataFrame
    melted_df = pd.melt(df, id_vars=id_columns, var_name='year', value_name='value')
    
    # Remove 'F' from the 'Year' column and filter non-numeric values
    melted_df['year'] = melted_df['year'].str.replace('F', '')
    melted_df = melted_df[melted_df['year'].str.isnumeric()]  # Filter out non-numeric values
    
    # Convert the 'Year' column to integers
    melted_df['year'] = melted_df['year'].astype(int)
    
    melted_df.name = melted_df_name

    # Save the reshaped DataFrame to a new CSV file
    # melted_df.to_csv(output_csv_filename, index=False)
    return melted_df 
```

```python
def save_dataframe_to_json(dataframe, filename):
    """
    Save a DataFrame to a JSON file.

    Args:
    dataframe (pd.DataFrame): The DataFrame to be saved.
    filename (str): The name of the JSON file.

    Returns:
    None
    """
    # Convert the DataFrame to a dictionary
    data_dict = dataframe.to_dict(orient='records')
    
    # Save the dictionary as a JSON file
    with open(filename, 'w') as json_file:
        json.dump(data_dict, json_file)
```
