```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
import numpy as np

reused_h2o_path = "data\\clean_data\\natural_reasources\\water_use\\Total_abstractions_of_reused_water.csv"
#read csv file into dataframe
reused_h2o_csv = pd.read_csv(reused_h2o_path)
```

```python
reused_h2o_csv.head()
```

```python
columns = reused_h2o_csv[['continent', 'year', 'h20_value']]

columns
```

```python
continent = 'Asia'

continent_data = reused_h2o_csv[reused_h2o_csv['continent'] == continent]

average_year = continent_data.groupby('year')['h20_value'].mean().reset_index()

average_year.columns = ['Year', 'Average_h20']


plt.figure(figsize=(10, 6))
plt.bar(average_year['Year'], average_year['Average_h20'], color='teal', alpha=0.6)
plt.xlabel('Year')
plt.ylabel('Average Water Abstraction (h20_value)')
plt.title(f'Average Water Abstraction Over the Years in {continent}')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("reused_water_asia.png")
plt.show()
```

```python
continent = 'Oceania'

continent_data = reused_h2o_csv[reused_h2o_csv['continent'] == continent]

average_year = continent_data.groupby('year')['h20_value'].mean().reset_index()

average_year.columns = ['Year', 'Average_h20']

plt.figure(figsize=(10, 6))
plt.scatter(average_year['Year'], average_year['Average_h20'], color='steelblue', alpha=0.6, label='Data Points')
slope, intercept, r_value, p_value, std_err = stats.linregress(average_year['Year'], average_year['Average_h20'])
line = slope * average_year['Year'] + intercept

plt.plot(average_year['Year'], line, color='red', label=f'Linear Regression (r-squared={r_value**2:.4f})')

plt.xlabel('Year')
plt.ylabel('Average Water Abstraction (h20_value)')
plt.title(f'Scatter Plot with Linear Regression in {continent}')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("reused_water_oceania_scat.png")
plt.show()
```
